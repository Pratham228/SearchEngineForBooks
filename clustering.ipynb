```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.cluster import KMeans
from sklearn.cluster import MiniBatchKMeans
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import json
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data = pd.read_json("C:/Users/prayaas/Documents/Courses/Information retrieval/full_data.json", lines=True)
data.head()
```

```python
# tfidf = TfidfVectorizer(
#     min_df = 5,
#     max_df = 0.95,
#     max_features = 8000,
#     stop_words = 'english'
# )
#v = CountVectorizer()

# cols = ['bookName', 'author', 'description']
# data['combined'] = data[cols].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
# data.shape



v1 = TfidfVectorizer(
    stop_words = 'english')
# # tokenize and build vocab
v1.fit(data.bookName)
# print(v1.idf_)
vector = v1.transform(data.bookName)
# print(vector.shape)
# #print(vector.toarray())

#v1 = TfidfVectorizer(stop_words = 'english')
```

```python
import time
start_time = time.time()
kmeans = KMeans(n_clusters=50, random_state=0).fit(vector)
print(time.time()-start_time)
```

```python
data["cluster_no"]=kmeans.labels_
data.head()
```
